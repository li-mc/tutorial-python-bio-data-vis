```python
import pandas as pd
import numpy as np

#Originally "http://figshare.com/download/file/2087487/1"
data = np.genfromtxt("https://ndownloader.figshare.com/files/3338657",
                     comments="#", usecols=range(2,73,1),
                     names=True, dtype=object, delimiter="\t")
df = pd.DataFrame(data)
print("Number of genes: " + str(len(df)))
df.head()
```

```python
#Convert byte arrays to strings.
str_df = df.stack().str.decode('utf-8').unstack()
```
